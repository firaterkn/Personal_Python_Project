```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast

df = pd.read_csv("C:\\Users\\ERKAN\\Downloads\\data_jobs.csv")
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)


df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df.job_title_short == "Data Analyst"].copy()
df_DA
```

```python
df_DA["job_posted_month_num"] = df_DA.job_posted_date.dt.month
df_DA
```

```python
df_DA_explode = df_DA.explode("job_skills")
df_DA_explode
```

```python
df_DA_explode_pivot = df_DA_explode.pivot_table(index="job_posted_month_num", columns="job_skills", aggfunc="size", fill_value=0)
```

```python
df_DA_explode_pivot.loc["Total"] = df_DA_explode_pivot.sum()
df_DA_explode_pivot
```

```python
df_DA_explode_pivot.loc["Total"].sort_values(ascending=False).index
```

```python
df_DA_explode_pivot = df_DA_explode_pivot[df_DA_explode_pivot.loc["Total"].sort_values(ascending=False).index]
df_DA_explode_pivot
```

```python
df_DA_explode_pivot = df_DA_explode_pivot.drop(index="Total")
df_DA_explode_pivot
```

```python
df_DA_explode_pivot.iloc[:, :5]
```

```python
df_DA_explode_pivot.iloc[:, :5].plot(kind="line", figsize=(10,6) )

plt.title("Top 5 Skills for Data Analyst per Months")
plt.ylabel("Count")
plt.xlabel("Months")
plt.legend()
plt.show()
```

```python
df_DA_pivot_2 = df_DA_explode_pivot.reset_index()
df_DA_pivot_2['job_posted_month'] = df_DA_pivot_2['job_posted_month_num'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%B'))
df_DA_pivot_2 = df_DA_pivot_2.set_index('job_posted_month')
df_DA_pivot_2 = df_DA_pivot_2.drop(columns='job_posted_month_num')

df_DA_pivot_2
```

```python
df_DA_pivot_2.iloc[:, :5].plot(kind='line', figsize=(10,6))

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('Months')
plt.legend()
plt.show()
```
