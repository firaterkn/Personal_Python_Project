```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:\\Users\\ERKAN\\Downloads\\data_jobs.csv")
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
```

```python
filt = df.salary_year_avg.notna()
df_filtered = df.loc[filt, "salary_year_avg"]
df_filtered
```

```python
def projected_salary(salary):
    return salary * 1.03

df_filtered_3_percent_high = df_filtered.apply(projected_salary)
df_filtered = df_filtered.to_frame()
df_filtered["df_3_percent_high"] = df_filtered_3_percent_high
df_filtered[["salary_year_avg", "df_3_percent_high"]]
```

```python
#Using lambda and apply
df_filtered["df_3_percent_high"] = df_filtered["salary_year_avg"].apply(lambda salary: salary * 1.03).notna()
df_filtered[["salary_year_avg", "df_3_percent_high"]]
```

```python
def inflation(salary):
    return salary * 1.03

df['salary_year_inflated'] = df['salary_year_avg'].apply(inflation)

df[pd.notna(df.salary_year_avg)][['salary_year_avg', 'salary_year_inflated']]
```

```python
#without apply method
df['salary_year_inflated'] = df['salary_year_avg'] * 1.03

df[pd.notna(df.salary_year_avg)][['salary_year_avg', 'salary_year_inflated']]
```

```python
    #Find inflated salaries but senior ones will have 5%, the others 3%

def projected_salary(row):
        if "Senior" in row["job_title_short"]:
            return  1.05 * row["salary_year_avg"]
        else:
            return  1.03 * row["salary_year_avg"]

df["salary_year_inflated"] = df.apply(projected_salary, axis=1)

df[pd.notna(df.salary_year_avg)][["job_title_short", "salary_year_avg", "salary_year_inflated"]]
```
