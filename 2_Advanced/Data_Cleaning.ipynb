```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:\\Users\\ERKAN\\Downloads\\data_jobs.csv")
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
```

```python
df.loc[0:10, ["salary_year_avg", "salary_hour_avg"]]
```

```python
year_avg = df.salary_year_avg.median()
hour_avg = df.salary_hour_avg.median()

df_filled = df

df_filled.salary_year_avg = df_filled.salary_year_avg.fillna(year_avg)
df_filled.salary_hour_avg = df_filled.salary_hour_avg.fillna(hour_avg)

df_filled.loc[0:10, ["salary_year_avg", "salary_hour_avg"]]
```

```python
df_drop_duplicate = df_filled
df_drop_duplicate = df_drop_duplicate.drop_duplicates()

print("Original length: ", len(df_filled))
print("Dropped data length: ", len(df_drop_duplicate))
print("Rows dropped: ", len(df_filled)-len(df_drop_duplicate))


```

```python
df_drop_duplicate = df_drop_duplicate.drop_duplicates(subset=["company_name", "job_title"])

print("Original length: ", len(df_filled))
print("Dropped data length: ", len(df_drop_duplicate))
print("Rows dropped: ", len(df_filled)-len(df_drop_duplicate))
```
