# How are in-demand skills trending for Data Analysts?

   ### Methology

   1. Aggregate skill counts monthly
   2. Re-analyze based on percentage of total jobs
   3. Plot the monthly skill demand


```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast
import seaborn as sns


# Load Data
df = pd.read_csv("C:\\Users\\ERKAN\\Downloads\\data_jobs.csv")

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Use just Data Analyst job titles in the United States, and copy not to change original dataframe
df_DA_US = df[(df.job_title == "Data Analyst") & (df.job_country == "United States")].copy()

# Add new column named job_posted_month_no, and get the month number in the job_posted_date column
df_DA_US["job_posted_month_no"] = df_DA_US.job_posted_date.dt.month

# Explode job_skills column to analyze and group them
df_DA_US_explode = df_DA_US.explode("job_skills")
```

```python
# Use pivot table to see all the details
df_DA_US_pivot = df_DA_US_explode.pivot_table(index="job_posted_month_no", columns="job_skills",  aggfunc="size", fill_value=0)
df_DA_US_pivot
```

```python
# Add a row to sum the column
df_DA_US_pivot.loc["Total"] = df_DA_US_pivot.sum()
df_DA_US_pivot
```

```python
# Sort from most frequent to least, then drop the Total column
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc["Total"].sort_values(ascending=False).index]
df_DA_US_pivot = df_DA_US_pivot.drop("Total")
df_DA_US_pivot
```

```python
# Take the top 5 and plot
df_DA_US_pivot.iloc[:, :5].plot(kind="line", figsize=(10,5))

plt.title("Trending Top Skills for Data Analysts in the US")
plt.ylabel("Counter")
plt.xlabel("Months")
plt.legend()
plt.show()
```

```python
# Take the number of Monthly Data Analyst Job Postings
DA_totals = df_DA_US.groupby("job_posted_month_no").size()
DA_totals
```

```python
# (Sum of the Job Skill in Month / Total Job Postings in Month) to find percentage
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)
```

```python
# Change from month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent["job_posted_month"] = df_DA_US_percent.job_posted_month_no.apply(lambda x: pd.to_datetime(x, format="%m").strftime("%b"))
df_DA_US_percent = df_DA_US_percent.set_index("job_posted_month")
df_DA_US_percent = df_DA_US_percent.drop(columns="job_posted_month_no")

df_DA_US_percent
```

```python

# Percent formatter for y axis
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, palette="tab10")
sns.set_theme(style="ticks")
sns.despine() # remove top and right spines

plt.title("Trending Top 5 Skills for Data Analysts in the US")
plt.ylabel("Percentage in Data Analyst Job Posting")
plt.xlabel("2023")
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color="black")

plt.show()
```
